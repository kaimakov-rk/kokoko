##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
#Агрегация
from typing import TypeAlias
from copy import deepcopy


DataFrame: TypeAlias = pd.core.frame.DataFrame


class Task:

  def __init__(self, dataframe: DataFrame):
    self._df = deepcopy(dataframe)

  @property
  def dataframe(self) -> DataFrame:
    return self._df

  def aggregate(self, category: str, numeric: str) -> dict:
    names: list = self._df.columns.values.tolist()
    def name_to_index(name: str, row: list) -> float:
      for i in range(len(names)):
        if name == names[i]:
          return i + 1
    def median(data: list) -> float:
      length: int = len(data)
      if length % 2 == 0:
        return(sum(sorted(data)[int(length / 2): int(length / 2) + 1]) / 2)
      else:
        return(sorted(data)[int(length / 2)])
    def mean(data: list) -> float:
      length = len(data)
      return sum(data) / length
    categories: dict = {value : list() for value in list(set([value for value in self._df[category]]))}
    self._df = self._df.reset_index()
    for index, row in self._df.iterrows():
      categories[row[name_to_index(category, row)]].append(row[name_to_index(numeric, row)])
    for key in categories.keys():
      categories[key] = (mean(categories[key]), median(categories[key]))
    return categories


task = Task(df)
print(task.aggregate("Type", "Process temperature [K]"))
```
